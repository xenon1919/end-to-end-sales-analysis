# Import necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv("../data/cleaned/sales_data_cleaned.csv")
```

```python
plt.style.use("default")
sns.set_theme()
```

### Business Questions
1. How does revenue vary over time?
2. Which product categories generate the most revenue?
3. Which regions perform best?
4. Does discount impact revenue?
5. Which customer segment is most valuable?


# Revenue Over Time

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
daily_revenue = df.groupby('OrderDate')['Revenue'].sum()

plt.figure(figsize=(10,4))
plt.plot(daily_revenue)
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

# Revenue by Product Category

```python
category_revenue = df.groupby('ProductCategory')['Revenue'].sum()

plt.figure(figsize=(6,4))
category_revenue.plot(kind='bar')
plt.title("Revenue by Product Category")
plt.ylabel("Revenue")
plt.show()
```

# Revenue by Region

```python
region_revenue = df.groupby('Region')['Revenue'].sum()

plt.figure(figsize=(6,4))
sns.barplot(x=region_revenue.index, y=region_revenue.values)
plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

# Discount vs Revenue

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Discount', y='Revenue')
plt.title("Discount vs Revenue")
plt.show()
```

# Customer Segment Analysis

```python
segment_revenue = df.groupby('CustomerSegment')['Revenue'].sum()

plt.figure(figsize=(6,4))
segment_revenue.plot(kind='pie', autopct='%1.1f%%')
plt.title("Revenue Share by Customer Segment")
plt.ylabel("")
plt.show()
```

# Correlation Heatmap

```python
numeric_df = df[['UnitPrice', 'Quantity', 'Discount', 'Revenue']]

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
